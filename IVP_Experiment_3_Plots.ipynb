```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_python = cv2.imread('python.png')
img_python_bw = cv2.imread('python.png',0)
cv2_imshow(img_python)
cv2_imshow(img_python_bw)
```

```python
minI = img_python_bw.min()
maxI = img_python_bw.max()
```

```python
min0=0
max0=100
k,l = img_python_bw.shape
new = np.zeros((k,l),dtype=np.uint8)
for i in range(k):
  for j in range(l):
    new[i,j]=(int)(img_python_bw[i,j]-minI)*(((max0-min0)/(maxI-minI))+min0)
cv2_imshow(new)
```

```python
hist = cv2.calcHist([img_python_bw],[0],None,[256],[0,256])
plt.plot(hist)
```

```python
hist2 = cv2.calcHist([new],[0],None,[256],[0,256])
plt.plot(hist2)
```

```python
equ = cv2.equalizeHist(img_python_bw)
hist = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.plot(hist)
```
