```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
sobel_mask_gx = np.array([[-1, -2, -1],
                          [0, 0, 0],
                          [1, 2, 1]])
sobel_mask_gy = np.transpose(sobel_mask_gx)
```

```python
hoover = cv2.imread("hoover.png", 0)
```

```python
vertical_hoover = cv2.filter2D(hoover, -1, sobel_mask_gx)
cv2_imshow(vertical_hoover)
```

```python
horizontal_hoover = cv2.filter2D(hoover,-1,sobel_mask_gy)
cv2_imshow(horizontal_hoover)
```
