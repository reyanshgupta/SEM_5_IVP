```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('nyc_skyline.png',0)
cv2_imshow(img)
```

```python
m,n=img.shape
import numpy as np
mask=np.ones([3,3],dtype=int)
mask=mask/9
img_new=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
    temp=img[i-1,j-1]*mask[0,0]+img[i-1,j]*mask[0,1]+img[i-1,j+1]*mask[0,2]+img[i,j-1]*mask[1,0]+img[i,j]*mask[1,1]+img[i,j+1]*mask[1,2]+img[i+1,j-1]*mask[2,0]+img[i+1,j]*mask[2,1]+img[i+1,j+1]*mask[2,2]
    img_new[i,j]=temp
cv2_imshow(img_new)
```

```python
img_new = img_new.astype(np.uint8)
cv2_imshow(img_new)
```

```python
blur = cv2.blur(img,(5,5))
cv2_imshow(img)
```

```python
mask=np.ones([5,5],dtype=int)
mask=mask/25
blur = cv2.blur(img,(5,5))
cv2_imshow(blur)
```

```python
res = np.hstack((img_new,blur))
cv2_imshow(res)
```

```python
res = np.hstack((img,img_new,blur))
cv2_imshow(res)
```

```python
blur1=cv2.blur(img,(8,8))
cv2_imshow(blur1)
```

```python
res=np.hstack((img,img_new,blur,blur1))
cv2_imshow(res)
```

```python
imgwithin = cv2.filter2D(img,-1,mask)
cv2_imshow(imgwithin)
```

```python
median = cv2.medianBlur(img,5)
cv2_imshow(median)
```

```python
mask = np.ones((5,5),np.uint8)
err = cv2.erode(img,mask)
cv2_imshow(err)
```

```python
dil = cv2.dilate(img,mask)
cv2_imshow(dil)
```

```python
res=np.hstack((img,dil,err))
cv2_imshow(res)
```

```python
bil = cv2.bilateralFilter(img,15,75,75)
cv2_imshow(bil)

blur = cv2.GaussianBlur(img,(5,5),0)
cv2_imshow(blur)
```
