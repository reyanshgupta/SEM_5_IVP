```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
from sympy import fwht
from sympy import ifwht
import sympy as sp
```

```python
hoover = cv2.imread('hoover.png')
hoover_bw=cv2.imread('hoover.png',0)
res_hoover = cv2.resize(hoover_bw, (0,0),fx=0.3,fy=0.3)
cv2_imshow(res_hoover)
res_hoover.shape
```

```python
reshaped_hoover = res_hoover.reshape(-1)
reshaped_hoover.shape
#converts 2D array to 1D
```

```python
x = reshaped_hoover.shape
x = int(x[0])
#last element of length(2d - length x breadth)
```

```python
#fast walsh hadamard transform
hadamard_hoover = fwht(reshaped_hoover)
```

```python
hadamard_hoover = hadamard_hoover[0:x]
```

```python
final_hadamard_hoover = np.reshape(hadamard_hoover,res_hoover.shape)
cv2_imshow(final_hadamard_hoover)
```

Inverse

```python
inverse_hoover = ifwht(hadamard_hoover)
```

```python
inverse_hoover = inverse_hoover[0:x]
```

```python
final_inverse_hoover = np.reshape(inverse_hoover,res_hoover.shape)
cv2_imshow(final_inverse_hoover)
```

```python
row,col=final_hadamard_hoover.shape
ec_hoover = final_hadamard_hoover
r=100
while(r<row):
  c=0
  while(c<col):
    ec_hoover[r][c] = 0
    c+=1
  r+=1

r=0
while(r<row):
  c=100
  while(c<col):
    ec_hoover[r][c] = 0
    c+=1
  r+=1
```

```python
cv2_imshow(ec_hoover)
ec_hoover_res = ec_hoover.reshape(-1)
```

```python
ec_inv = ifwht(ec_hoover_res)
ec_inv = ec_inv[0:x]
final_ec_inv = np.reshape(ec_inv,res_hoover.shape)
cv2_imshow(final_ec_inv)
```
