```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('nyc_skyline.png',0)
cv2_imshow(img)
```

```python
kernel = np.ones((5,5),np.uint8)
img_erosion = cv2.erode(img,kernel,iterations=1)
img_dilation=cv2.dilate(img,kernel,iterations=1)

cv2_imshow(img)
cv2_imshow(img_erosion)
cv2_imshow(img_dilation)

cv2.waitKey(0)
```

```python
mask=np.ones((11,11),np.uint8)
dil=cv2.dilate(img,mask)
cv2_imshow(dil)

mask=np.ones((15,15),np.uint8)
dil1=cv2.dilate(img,mask)
cv2_imshow(dil1)

mask=np.ones((45,45),np.uint8)
dil3=cv2.dilate(img,mask)
cv2_imshow(dil3)



res=np.hstack((img,dil,dil1,dil3))
cv2_imshow(res)
```

```python
T=100
m,n=img.shape
img_thresh=np.zeros((m,n),dtype=np.uint8)
for i in range(m):
  for j in range(n):
    if img[i,j]<T:
      img_thresh[i,j]=0
    if img[i,j]>T:
      img_thresh[i,j]=255
cv2_imshow(img_thresh)
```

```python
ress=cv2.morphologyEx(img_thresh,cv2.MORPH_HITMISS,mask)
cv2_imshow(ress)
```
