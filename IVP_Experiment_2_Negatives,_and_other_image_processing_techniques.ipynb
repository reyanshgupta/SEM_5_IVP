```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt1
```

```python
img_python_bw = cv2.imread('python.png',0)
```

```python
cv2_imshow(img_python_bw)
```

**Negative Image:**

```python
m,n = img_python_bw.shape
img_python_neg = img_python_bw
for i in range(m):
  for j in range(n):
    img_python_neg[i,j] = 255-img_python_bw[i,j]
cv2_imshow(img_python_neg)
```

**Threshold**

```python
img_python_thresh = np.zeros((m,n),dtype=int)
for i in range (m) :
  for j in range (n) :
    if(img_python_bw[i,j]>50):
      img_python_thresh[i,j]=255
    else:
      img_python_thresh[i,j]=0
cv2_imshow(img_python_thresh)
```
